```python
df['institute_type'] = [0 if x =='IIT' else 1 for x in df['institute_type']]
df['institute_type']
```

```python
df['quota'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['quota'] = le.fit_transform(df['quota'])
df['quota'].value_counts()
```

```python
df['pool'].value_counts()
```

```python
df['pool'] = [0 if x == 'Gender-Neutral' else 1  for x in df['pool']]
df['pool'].value_counts()
```

```python
df['program_duration'].value_counts()
```

```python
df['program_duration'] = [0 if x == '4 Years' else 1 for x in df['program_duration']]
df['program_duration'].value_counts()
```

```python
df['degree_short'].value_counts()
```

```python
df['degree_short'] = le.fit_transform(df['degree_short'])
df['degree_short'].value_counts()
```

```python
df['category'].value_counts()
```

```python
df['category'] = le.fit_transform(df['category'])
df['category'].value_counts()
```

```python
df.head()
```

```python
X = df.drop(columns='is_preparatory')
X
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
acc = accuracy_score(y_test, y_pred.round())
```
