```python
# Dropping unnecessary columns

clg_data.drop(columns=['id','institute_short','program_name'], inplace=True)
```

```python
clg_data.head()
```

```python
clg_data.shape
```

    After removing we have 11 columns 

```python
clg_data.columns
```

```python
# Checking the types of columns we have
clg_data.info()
```

```python
# Converting all Object Columns to number for ease of fitting model

print('Finding the types of values in institute type: ',clg_data['institute_type'].value_counts())
```

```python
clg_data['institute_type'] = [0 if x =='IIT' else 1 for x in clg_data['institute_type']]
clg_data['institute_type']
```

```python
clg_data['pool'] = [0 if x == 'Gender-Neutral' else 1  for x in clg_data['pool']]
clg_data['pool'].value_counts()
```

```python
clg_data['program_duration'] = [0 if x == '4 Years' else 1 for x in clg_data['program_duration']]
clg_data['program_duration'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
clg_data['quota'] = le.fit_transform(clg_data['quota'])
clg_data['quota'].value_counts()
```

```python
clg_data['degree_short'] = le.fit_transform(clg_data['degree_short'])
clg_data['degree_short'].value_counts()
```

```python
clg_data['category'] = le.fit_transform(clg_data['category'])
clg_data['category'].value_counts()
```

```python
clg_data.head()
```

    Splitting the data into Independant and Dependant Variable

```python
X = clg_data.drop(columns='is_preparatory')
y = clg_data.iloc[:,-1]
```

    Spltting the data into Trainig and Testing dataset
   

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Creating the object of Logistic Regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
# Testing using test dataset
y_pred = lr.predict(X_test)
```

```python
# Finding accuracy of model
from sklearn.metrics import accuracy_score, classification_report

acc = accuracy_score(y_test, y_pred.round())
acc
```

    Model accuracy is 95%

```python
# Creating a pickle file for further predictions

import joblib
joblib.dump(lr,'college admission predictor.pkl')
```
